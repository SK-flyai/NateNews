```python
import re
```

```python
tmp = """
[https://asdfasdf] hihi Hi? [asdfadsf]
asdfa asdfa? asdjfhasljk [https:// asdaj?] asdf[12341234] [https://asdlfkja?asdfl!]
"""

pattern_link = re.compile('\[(https://[^\]]*)\]')
result_img = pattern_link.finditer(tmp)
tmp = re.sub(pattern_link, '', tmp)

pattern_cap = re.compile('\[([^\]]*)\]')
result_cap = pattern_cap.finditer(tmp)
tmp = re.sub(pattern_cap, '', tmp)

print(tmp)
```

```python
for r in result_img:
    print(r.group(1))
```

```python
result_list = dict()
for r in result_img:
    result_list[r.group(1)] = ''
    try:
        result_list[r.group(1)] = next(result_cap).group(1)
    except:
        print('no caption!')
```
